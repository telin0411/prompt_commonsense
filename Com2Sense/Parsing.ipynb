```python
class parsingTreeNode():
    def __init__(self, parsingTree_node):
        self.type = ''
        self.data = ''
        self.parent = parsingTree_node
        self.children = []
    
    def add_child(self, parsingTree_node):
        self.children.append(parsingTree_node)
        
    def add_data(self, node_data: str):
        self.data = node_data
        
    def add_type(self, node_type: str):
        self.type = node_type

        
def createTree(tree_str: str):
    
    root = parsingTreeNode(None)
    current_node = root
    
    for char in tree_str:
        
        char_cache = ''
        
        if char == '(':
            new_node = parsingTreeNode(current_node)
            current_node.add_child(new_node)
            current_node = current_node.children[-1]
        
        elif char == ')':
            current_node = current_node.parent
            
        elif char == ' ':
            if current_node.type == '':
                current_node.add_data(char_cache)
            else:
                current_node.add_data(char_cache)
            char_cache = ''
        else:
            char_cache += char
    return root
        
```
